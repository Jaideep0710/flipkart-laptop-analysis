```python
import requests
import pandas as pd
import time
import random
from bs4 import BeautifulSoup

# Basic setup
URL = "https://www.flipkart.com/search?q=laptop&sort=price_desc&page={}"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
Total_Page = 240
Output_file = "flipkart_laptop.csv"

# store all laptop data
all_Laptops = []

def laptop_page(page_num):
    try:
        url = URL.format(page_num)
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"Failed to fetch page {page_num}")
            return []
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return []

    soup = BeautifulSoup(response.text, "html.parser")
    data = []

    # Product Cards
    product_card = soup.find_all('div', {'class': "tUxRFH"})
    for card in product_card:
        # Name
        name_tag = card.find("div", {"class": "KzDlHZ"})
        name = name_tag.text.strip() if name_tag else None

        # Price
        price_tag = card.find("div", {'class': 'Nx9bqj _4b5DiR'})
        price = price_tag.text.strip().replace("₹", "").replace(",", "") if price_tag else None

        # Discount
        discount_tag = card.find("div", {'class': "UkUFwK"})
        discount = discount_tag.text.strip() if discount_tag else None

        # Original Price
        orginal_price = card.find('div', {'class': 'yRaY8j ZYYwLA'})
        orginal = orginal_price.text.strip().replace("₹", "").replace(",", "") if orginal_price else None

        # Product Left
        product_left = card.find('div', {'class': 'M4DNwV'})
        prod_left = product_left.text.strip() if product_left else None

        # Description
        desc_contain = card.find('div', {"class": "_6NESgJ"})
        if desc_contain:
            desc_list = desc_contain.find_all('li', {"class": "J+igdf"})
            desc = '; '.join([d.text.strip() for d in desc_list])
        else:
            desc = None

        if name:
            data.append({
                "Name": name,
                "Price": price,
                "Discount": discount,
                "Original_price": orginal,
                "Description": desc,
                "Product_left": prod_left
            })

    return data


# Main loop
for page in range(1, Total_Page + 1):
    print(f"Scraping page {page}/{Total_Page}")
    products = laptop_page(page)
    all_Laptops.extend(products)

    # Save progress every 20 pages
    if page % 20 == 0:
        pd.DataFrame(all_Laptops).to_csv(Output_file, index=False)
        print(f"✅ Progress saved at page {page}")

    # Sleep to avoid blocking
    time.sleep(random.uniform(1, 3))

# Final save
df = pd.DataFrame(all_Laptops)
df.to_csv(Output_file, index=False)
print(f"\n✅ Done! Collected {len(df)} products into {Output_file}")
```
